```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import html5lib
from functions import *
```

```python
formation_stats = pd.read_csv("psg_formations.csv")
formation_stats.xG = formation_stats.xG.apply(lambda xG : correctValue(xG))
formation_stats.xGA = formation_stats.xGA.apply(lambda xGA : correctValue(xGA))
formation_stats["G90"] = formation_stats.G / (formation_stats.Min/90)
formation_stats.G90 = formation_stats.G90.apply(lambda G90 : "{:.2f}".format(G90))
formation_stats
```

```python
plt.pie(formation_stats.xG90,labels=formation_stats.Formation,autopct='%1.0f%%')
plt.title("Expected Goals per Match for each formation")
plt.show()
```

```python
plt.pie(formation_stats.xGA90,labels=formation_stats.Formation,autopct='%1.0f%%')
plt.title("Expected Goals Against per Match for each formation")
plt.show()
```

```python
plt.pie(formation_stats.G90,labels=formation_stats.Formation,autopct='%1.0f%%')
plt.title("Average Goals Per Matche For each formation")
plt.show()
```

```python
plt.barh(formation_stats.Formation,formation_stats.xG90-formation_stats.xGA90)
plt.title("Difference between \"for\" and \"against\" expected goals for each formation")
plt.show()
```
